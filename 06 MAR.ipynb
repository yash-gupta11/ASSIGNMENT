QUESTION 5
ANSWER : Concept Measurement of Data : In this concept, we have some condition based on the property of data to be analyse.So that we can find the best summary and insights from the data 
            There are 4 level of measurement of data 
                (i) Nominal scale data : 
                    (a) Categorical without rank,
                    ex : genter ,blood group etc.
                (ii) Ordinance scale data : 
                    (a) Categorical without rank
                    (b) Difference not allowed 
                    ex : Rank of student in class,customer feedback etc.
                (iii) Interval scale data :
                    (a) order matters.
                    (b) differnce allowed.
                    (c) ratio not allowed.
                    ex : temperatue variable in 'f' and temp != 0.
                (iv) Ratio scale data : 
                    order matters.
                    (b) differnce allowed.
                    (c) ratio allowed.
                    ex : marks of student in class etc.
