```python
#method 1
fl = sns.load_dataset('flights')
fl.drop(labels = 'month',axis = 1,inplace = True)
sns.heatmap(fl)
```

```python
#method 2
fl1 = sns.load_dataset('flights')
def mon(x):
    if x == 'Jan':
        return 1
    elif x == 'Feb':
        return 2
    elif x == 'Mar':
        return 2
    elif x == 'Apr':
        return 4
    elif x == 'May':
        return 5
    elif x == 'Jun':
        return 6
    elif x == 'Jul':
        return 7
    elif x == 'Aug':
        return 8
    elif x == 'Sep':
        return 9
    elif x == 'Oct':
        return 10
    elif x == 'Nov':
        return 11
    elif x == 'Dec':
        return 12
fl1['month'] = fl1['month'].apply(mon)
```

```python
sns.heatmap(fl1)
```
