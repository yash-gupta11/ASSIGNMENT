QUESTION 2

```python
#The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.show()
```

```python
# subplot is allow as the abality to produce one or more graph in single show.
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)

plt.show()
```

```python
#The bar() function takes arguments that describes the layout of the bars.The categories and their values represented by the first and second argument as arrays.
# Bar graph is used to represent the categorical data and number array.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.show()

plt.barh(company,profit)
plt.show()
```
